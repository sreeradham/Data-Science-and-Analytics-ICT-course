Who scores the most on average for math, reading and writing based on
 Gender  Test preparation course


```python
#create subplots

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Score')
plt.subplot(1,3,2)
sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Reading Score')
plt.subplot(1,3,3)
sns.boxplot(x='gender', y='writing score', data=df)
plt.title('Writing Score')
plt.show()
```

The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this

```python
top_25_percentile = df['math score'].quantile(0.75)
df['bonus_points'] = df['math score'].apply(lambda x: 5 if x >= top_25_percentile else 0)

```

```python
# Let's print the updated dataframe with the bonus points column
print("\nUpdated DataFrame with bonus points:")
print(df)
```
